```python
# Instalar bibliotecas necessÃ¡rias
!pip install rembg pillow matplotlib

# ImportaÃ§Ãµes
import os
from rembg import remove
from PIL import Image
from google.colab import files
import matplotlib.pyplot as plt

# Criar pastas de entrada e saÃ­da
os.makedirs("entrada", exist_ok=True)
os.makedirs("saida", exist_ok=True)

# Upload manual das imagens
print("ðŸ“‚ FaÃ§a upload das imagens que deseja processar:")
uploaded = files.upload()

# Salvar imagens enviadas na pasta "entrada"
for nome, conteudo in uploaded.items():
    with open(os.path.join("entrada", nome), 'wb') as f:
        f.write(conteudo)

# Listar imagens encontradas
arquivos = os.listdir("entrada")
print("Imagens encontradas:", arquivos)

# Processar imagens (remover fundo)
for nome in arquivos:
    caminho_in = os.path.join("entrada", nome)
    caminho_out = os.path.join("saida", nome)

    with open(caminho_in, "rb") as inp:
        input_data = inp.read()
        output_data = remove(input_data)

    with open(caminho_out, "wb") as out:
        out.write(output_data)

print("âœ… Todas as imagens foram processadas e salvas na pasta 'saida'.")

# Teste de validaÃ§Ã£o
assert len(os.listdir("saida")) > 0, "Nenhuma imagem foi salva!"
print("ðŸŸ¢ Teste passou: imagens salvas corretamente.")

# Mostrar a primeira imagem processada
primeira_img = os.listdir("saida")[0]
caminho_img = os.path.join("saida", primeira_img)

img = Image.open(caminho_img)
plt.imshow(img)
plt.axis("off")
plt.show()

```
